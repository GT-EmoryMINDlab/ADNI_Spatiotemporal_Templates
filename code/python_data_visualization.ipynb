```python
import os
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from scipy.io import loadmat
from nilearn import datasets
from nilearn.plotting import plot_stat_map
from nilearn.image import new_img_like, resample_to_img
import cv2

# --- Init ---
qpp_cohort = "ADNI_DAT_QPP"

def ensure_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# --- Load the probabilistic ICN atlas ---
atlas_path = "Neuromark_fMRI_2.2_modelorder-multi.nii"  # Replace with actual file
atlas_img = nib.load(atlas_path)
atlas_data = atlas_img.get_fdata()

# --- Load QPP template from .mat file ---
qpp_mat = loadmat(f"{qpp_cohort}.mat")  # Replace with actual file
qpp_signal = qpp_mat[qpp_cohort]  # Adjust this to match the actual variable name

# --- Load and resample MNI152 Template ---
t1_img = datasets.load_mni152_template()
resampled_t1 = resample_to_img(t1_img, atlas_img, interpolation="nearest")
resampled_mask = resampled_t1.get_fdata() > 0

# --- Compute Weighted Sums for All Time Points ---
weighted_brains = np.array([
    np.sum(atlas_data * qpp_signal[:, t], axis=3) for t in range(qpp_signal.shape[1])
])

# --- Normalize to range [-1, 1] ---
global_min, global_max = np.min(weighted_brains), np.max(weighted_brains)
if global_max - global_min > 0:
    weighted_brains = 2 * (weighted_brains - global_min) / (global_max - global_min) - 1
else:
    weighted_brains[:] = 0

# --- Apply Mask ---
masked_weighted_brains = np.where(resampled_mask, weighted_brains, np.nan)

# --- Save Images ---
output_dir = "interm"
ensure_directory(output_dir)

for t in range(qpp_signal.shape[1]):
    print(f"Processing time {t + 1}/{qpp_signal.shape[1]}")
    weighted_img = new_img_like(atlas_img, masked_weighted_brains[t])
    fig, ax = plt.subplots(figsize=(6, 3))
    plt.axis('off')
    plot_stat_map(weighted_img, title=f"QPP Signal at t={t}", display_mode='ortho',
                  cut_coords=(0, 0, 0), cmap='jet', figure=fig, vmax=1)
    img_path = os.path.join(output_dir, f"frame_{t:04d}.png")
    plt.savefig(img_path)
    plt.close(fig)

# --- Create Video ---
def create_video(image_folder, video_filename, fps=5):
    images = sorted([img for img in os.listdir(image_folder) if img.endswith(".png")])
    if not images:
        print("No images found for video creation.")
        return

    first_image = cv2.imread(os.path.join(image_folder, images[0]))
    height, width, _ = first_image.shape
    video_writer = cv2.VideoWriter(video_filename, cv2.VideoWriter_fourcc(*"mp4v"), fps, (width, height))

    for image in images:
        img_path = os.path.join(image_folder, image)
        frame = cv2.imread(img_path)
        video_writer.write(frame)

    video_writer.release()
    print(f"Video saved as {video_filename}")

# Generate video
video_filename = f"{qpp_cohort}_visualization.mp4"
create_video(output_dir, video_filename, fps=3)
```
